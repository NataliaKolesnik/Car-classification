В качестве экспериментов были использованы разные:

   - предобученные модели (EfficientNetV2M,  EfficientNetB5, EfficientNetB6, Xception)
   - разрешение (160х200, 240х300)
   - аугментация данных (albumentations и встроенная в keras)
   
Структура нейросети была одинаковая.

Попробуем теперь файлы submission собрать в один датасет и большинством голосов (moda) решить по каждому объекту что это за модель.

```python
sub_0 = pd.read_csv('submission/submission_0.csv')
sub_1 = pd.read_csv('submission/submission_1.csv')
sub_2 = pd.read_csv('submission/submission_2.csv')
sub_3 = pd.read_csv('submission/submission_3.csv')
sub_4 = pd.read_csv('submission/submission_4.csv')
sub_5 = pd.read_csv('submission/submission_5.csv')
```

```python
sub_0 = sub_0.rename(columns={'Category': 'Cat_0'})
sub_1 = sub_1.rename(columns={'Category': 'Cat_1'})
sub_2 = sub_2.rename(columns={'Category': 'Cat_2'})
sub_3 = sub_3.rename(columns={'Category': 'Cat_3'})
sub_4 = sub_4.rename(columns={'Category': 'Cat_4'})
sub_5 = sub_5.rename(columns={'Category': 'Cat_5'})
```

```python
submission = sub_0.merge(sub_1, on='Id', how='left')
submission = submission.merge(sub_2, on='Id', how='left')
submission = submission.merge(sub_3, on='Id', how='left')
submission = submission.merge(sub_4, on='Id', how='left')
submission = submission.merge(sub_5, on='Id', how='left')
```

```python
submission
```

```python
submission['Category'] = submission.apply(lambda row: statistics.mode([row['Cat_0'], row['Cat_1'], 
                                                                       row['Cat_2'], row['Cat_3'],
                                                                       row['Cat_4'], row['Cat_5']]), axis=1)
submission
```

```python
submission = submission[['Id','Category']]
```

```python
submission.to_csv('submission/submission_final.csv', index=False)
```
